```python
class From(object):
    conditionals = []
    cpt_table = [
	{
		"name": "play",
		"probabilities": [
			{
				"parent": "",
				"parentState": "",
				"values": {
						"yes": 0.64,
						"no": 0.36
				}
			}
		]
	},
	{
		"name": "outlook",
		"probabilities": [
			{
				"parent": "play",
				"parentState": "yes",
				"values": {
						"sunny": 0.14,
						"overcast": 0.28,
						"rainy": 0.21
				}
			},
			{
				"parent": "play",
				"parentState": "no",
				"values": {
						"sunny": 0.21,
						"overcast": 0.08,
						"rainy": 0.08
				}
			}
		]
	}
]
    #class_var = 'play'
    
    def __init__(self, bayesian_network):
        self.bayesian_network = bayesian_network
    
    def given(self, name, value):
        self.conditionals.append({"name": name, "value": value})
        return self
    
    def compute_probabilities(self, class_attribute_name):
        result_sum = 1
        
        for i, conditional in enumerate(self.conditionals):
            for j, node in enumerate(self.bayesian_network): 
                if (node['name'] == conditional['name']):
                    result_sum *= self.find_probability_in_cpt_table(node['name'], self.find_conditional_value(node['name']), node['parent'], self.find_conditional_value(node['parent']))
                    
        return result_sum

    def find_conditional_value(self, name):
        for cond in self.conditionals:
            if (cond['name'] == name):
                return cond['value']

    def find_probability_in_cpt_table(self, node_name, node_state, parent_name, parent_state):
        for cpt in self.cpt_table:
            if (cpt['name'] == node_name):
                for probability in cpt['probabilities']:
                    if (parent_name == "" or parent_state == None):
                        return probability["values"][node_state]
                        
                    if (probability['parent'] == parent_name and probability['parentState'] == parent_state):
                        return probability["values"][node_state]
        return 1
    
    def find_class_attribute_states(self, class_attribute_name):
        for node in enumerate(self.bayesian_network):
            if (node['name'] == class_attribute_name):
                return node['r']
            
        print("No class attribute state Error")
        return []
            
```

```python
network = [ 
	{'r': ['yes', 'no'], 'q': ['parent'], 'name': 'play', 'parent': ''},
	{'r': ['sunny', 'overcast', 'rainy'], 'q': ['yes', 'no'], 'name': 'outlook', 'parent': 'play'}
]

result = From(network).given('outlook', 'sunny').compute_probabilities('play')

print (result)
```
